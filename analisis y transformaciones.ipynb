```python
import pandas as pd 
import numpy as np

df=pd.read_csv('AccidentesAviones.csv')
```

```python
#se analizan las columnas 

df.head()
```

```python
#se reemplazan los nulos (?) por Nan para tenerlos en un formato mas general
df=df.replace('?',np.nan)
```

```python
#despues de analizar el dataset llegue a la conclusion que no necesito algunas columnas 
df=df.drop(columns=['flight_no','registration','cn_ln'])
```

```python
#vemos como quedo el dataset
df.head(20)
```

```python
#hay muchos datos nulos en la columna que expresa la mortalidad, se procede a elimnar estos datos ya que el estudio trata de la tasa de esta. 
df=df.dropna(subset='cantidad de fallecidos')
df=df.dropna(subset='passenger_fatalities')
```

```python
#vemos de nuevo el datset
df.head(30)
```

```python
#al ver que hay saltos en el indice del dataset le hacemos un pequeño recet
df=df.reset_index(drop=True)
```

```python
#se debe arreglar la columna de fechas para poder usar estos datos
df['fecha'] = pd.to_datetime(df['fecha'], format='%B %d, %Y').dt.strftime('%Y-%m-%d')

```

```python
#vemos como quedo el dataset con estos cambios
df.head()
```

```python
#al parecer hay más columna que no parecen aportar al analisis por lo tanto procedere a eliminarlas 
df=df.drop(columns=['summary','route'])
```

```python
#al aplicar la función me doy cuenta que hay caracters de mas en la columna tratare de arreglarla filtrando los datos y normalizando la columna 
df[df['HORA declarada'].str.len() > 4]
#veo los datos en excel para tener saber que datos debo reemplezar en python
df['HORA declarada'] = df['HORA declarada'].str.replace('c', '').str.replace('Z','').str.replace(';', '').str.replace(':','')
```

```python
#vuelvo a filtrar
df[df['HORA declarada'].str.len()>4]
```

```python
#voy a darle formato a la columna, al darle el formato de la columna da un error por espacios precedere a elimnarlos usando el comando strip
df['HORA declarada']=df['HORA declarada'].str.strip()
df['HORA declarada']=pd.to_datetime(df['HORA declarada'], format='%H%M').dt.time


```

```python
df.head()
```

```python
#la columna pasajeros abordo tiene vacios los cuales no tengo forma de llena por lo cual los voy a eliminar
df=df.dropna(subset='all_aboard')
```

```python
#Por ultimo cambiare los nombres de las columnas para tener un formato mas uniforme
df.rename(columns={'HORA declarada':'hora_declarada','OperadOR':'operador','ac_type':'tipo_avion','all_aboard':'total_abordo','PASAJEROS A BORDO':'pasajeros',
 'crew_aboard':'tripulacion','cantidad de fallecidos':'total_fallecidos','passenger_fatalities':'pasajeros_fallecidos',
 'crew_fatalities':'tripulantes_fellecidos','ground':'fallecidos_tierra'},inplace=True)
```

```python
#ultimo reset antes de guardar el nuevo dataset
df=df.reset_index(drop=True)
```

```python
df.head()
```

```python
df.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
```

```python
df.head()
```

```python
total_vuelos = df.shape[0]
print(f'El número total de vuelos es: {total_vuelos}')
```

```python
#columna pra kpi
df['total_fallecidos'] = df['total_fallecidos'].astype(int)
df['fallecidos'] = df['total_fallecidos'].apply(lambda x: 'Sí' if x > 0 else 'No')
```

```python
#columna para kpi
tasa_accidentes = (
    df.groupby('Ruta')
    .agg(
        total_accidentes_ruta=('total_fallecidos', lambda x: (x > 0).sum()),
        total_vuelos_ruta=('Ruta', 'count')
    )
    .assign(tasa_accidentes=lambda x: x['total_accidentes_ruta'] / x['total_vuelos_ruta'])*100
)
df = pd.merge(df, tasa_accidentes, on='Ruta', how='left')
```

```python
#columna para kpi
tasa_accidentes_operador = (
    df.groupby('operador')
    .agg(
        total_accidentes_operador=('total_fallecidos', lambda x: (x > 0).sum()),
        total_vuelos_operador=('operador', 'count')
    )
    .assign(tasa_accidentes=lambda x: x['total_accidentes_operador'] / x['total_vuelos_operador'])*100
)

df = pd.merge(df, tasa_accidentes_operador, on='operador', how='left')
```

```python
#kpi
df['total_abordo'] = pd.to_numeric(df['total_abordo'], errors='coerce')

df['tasa_mortalidad'] = (df['total_fallecidos'] / df['total_abordo']) * 100
```

```python
#kpi
df['supervivientes'] = df['total_abordo'] - df['total_fallecidos']
df['tasa_supervivientes'] = (df['supervivientes'] / df['total_abordo']) * 100
```

```python
#aqui se exporta el archivo csv
df.to_csv('df.csv',index=False)
```
